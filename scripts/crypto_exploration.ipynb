```python
import sys
from io import BytesIO
import pandas as pd
import boto3
from botocore.client import Config

# 1. Define your MinIO connection parameters directly
MINIO_ENDPOINT = "http://localhost:9000"
MINIO_USER = "minicrypto"
MINIO_PASSWORD = "cryptopassword123"
MINIO_REGION = "us-east-1"

# 2. Initialize the S3 Client
s3_client = boto3.client(
    "s3",
    endpoint_url=MINIO_ENDPOINT,
    aws_access_key_id=MINIO_USER,
    aws_secret_access_key=MINIO_PASSWORD,
    config=Config(signature_version="s3v4"),
    region_name=MINIO_REGION
)

# 3. Specify your silver Parquet file key (Update this to match your exact file name if needed)
bucket_name = "crypto-silver"
file_key = "year=2026/month=06/day=09/coingecko_snapshot_140149.parquet"

print(f"🔄 Pulling clean data from: {bucket_name}/{file_key}...")

try:
    # 4. Download file from MinIO into memory
    obj = s3_client.get_object(Bucket=bucket_name, Key=file_key)
    
    # 5. Read the Parquet byte stream into a Pandas DataFrame
    df_silver = pd.read_parquet(BytesIO(obj['Body'].read()))
    print("✅ Data successfully loaded into Jupyter Memory!")
    
except Exception as e:
    print(f"❌ Error loading file. Did you copy the file key exactly? Details: {e}")

# 6. Display the data structure
df_silver.head(10)
```
